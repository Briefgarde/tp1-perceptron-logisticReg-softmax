### Effect of luck and random seed on accuracy

Doing this little test abouve lets us see the effect of luck and different random seeds on the model. While none of the different seeds led to any drastic improvement or regression of the accuracy, they typically make the final best accuracy of the model (which uses the best hyperparameter we found earlier) oscilliate by about -1.5/+1.5 pourcentage point around the mean.  

This randomness comes from how the weights are initialized in the model, and how batches are then selected during the training process.
